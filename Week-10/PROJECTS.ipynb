```python
class equation():
    
    def trapezium(self,height,base1,base2):
        height = int(input("Enter a height:"))
        base1 = int(input("Enter Base1:"))
        base2 = int(input("Enter Base2:"))
        self.height = height
        self.base1 = base1
        self.base2 = base2
        area = self.height / 2 * (self.base1 + self.base2)
        print(area)
        
    def rhombus(self,diagonal1,diagonal2):
        diagonal1 = int(input("Enter Diagonal1"))
        diagonal2 = int(input("Enter Diagonal2"))
        self.diagonal1 = diagonal1
        self.diagonal2 = diagonal2
        area = 0.5 * self.diagonal1 * self.diagonal2
        print(area)
        
    def parallelogram(self,base,altitude):

        base = int(input("Enter base:"))
        altitude = int(input("Enter altitude"))
        self.base = base
        self.altitude = altitude
        area = base * altitude
        print(area)
        
    def cube(self,length):
        length = int(input("Enter length"))
        self.length = length
        area = 6 * (length)**2
        print(area)
        
    def volume_of_cylinder(self,radius,height):
        radius = int(input("Enter radius"))
        height = int(input("Enter height"))
        self.radius = radius
        self.height = height
        area = 3.14 * radius**2 * height
        print(area)






solve = equation()
print(" 1. Area of Trapezium \n 2. Area of Rhombus \n 3. Area of Parallelogram \n 4. Area of Cube formula \n 5. Volume of Cylinder ")
sd = int(input("Enter what you want to solve:"))
if sd == 1:
    solve.trapezium(height,base1,base2)
elif sd == 2:
    solve.rhombus(diagonal1,diagonal2)
elif sd == 3:
    solve.parallelogram(base,altitude)
elif sd == 4:
    solve.cube(length)
elif sd == 5:
    solve.volume_of_cylinder(radius,height)
    
    
        
        
```
